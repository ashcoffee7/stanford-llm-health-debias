```python
allUrls = []
def prepare_prompt(open_prompt, closed_prompt):
  tokens_limit = 1430 # Limit for gpt-3.5-turbo
  # build our prompt with the retrieved contexts included

  count_of_tokens_consumed = len(encoding.encode(closed_prompt)) + len(encoding.encode(open_prompt))

  count_of_tokens_for_context = tokens_limit - count_of_tokens_consumed

  contexts = ""

  prompt_embedding = get_embedding(open_prompt, engine=embedding_model)
  df["similarity"] = df.embedding.apply(lambda x: cosine_similarity(x, prompt_embedding))
  results = (df.sort_values("similarity", ascending=False))
  print(results)
  print(results.head(3))
  x = results.head(3)



  # Fill in context as long as within limit
  for i in range(len(results)):
    if (count_of_tokens_for_context>=results.n_tokens.iloc[i]):
        contexts += results.text.iloc[i] + "\n"
        count_of_tokens_for_context -=1
        count_of_tokens_for_context -= results.n_tokens.iloc[i]

  complete_prompt = closed_prompt + contexts + open_prompt
  return complete_prompt


```

```python
# functions to set up closed prompts

def create_closed(i, dose):
  p = "Vignette: " + final_list[i][0] + "\n" + \
  "Question: " + final_list[i][1] + "\n" + \
  "Answer: " + final_list[i][2] + "\n" + \
  dose + "\n" + \
  "Explanation: " + final_list[i][4] + "##\n\n"
  return p

def standardize_closed(p, patient):
  p = p.replace('[gender] ', '')
  p = p.replace('[race] ', '')
  p = p.replace('[possessive]', 'their')
  p = p.replace('[subject]', patient)
  p = p.replace('Patient D', patient)
  return p

# functions to set up open prompts
def create_open_standard(q):
  p = "Vignette: " + data[q][0] + "\n" + \
  "Question: " + data[q][1] + "\n" + \
  "Answer:"
  return p

# functions to add race, gender, sexuality, and religion to patient profiles in vignettes
# adding gender
def genderize_open(p, g):
  p = p.replace('[gender]', g)
  p = p.replace('[possessive]', pronouns['possessive'][g])
  p = p.replace('[subject]', pronouns['subject'][g])
  return p

# adding gender, deleting race
def genderize_norace_open(p,g):
  p = p.replace('[gender]', g)
  p = p.replace('[possessive]', pronouns['possessive'][g])
  p = p.replace('[subject]', pronouns['subject'][g])
  p = p.replace('[race]', '')
  return p

# adding race
def race_name_open(p, r, q):
  p = p.replace('[race]', r)
  p = p.replace('Patient D', names[r][g][q])
  return p

# adding race with no associated name
def race_noname_open(p, r):
  p = p.replace('[race]', r)
  return p

# adding race, deleting gender
def race_nogender_open(p, r, q):
  p = p.replace('[gender]', "individual")
  p = p.replace('[race]', r)
  g = random.randrange(0,2)
  return p

# adding sexuality, deleting race
def sexuality_open(p, s):
  p = p.replace('[gender]', s + '-identifying patient')
  p = p.replace('[race]', '')
  return p

# combining sexuality with gender
def sexuality_open_g(p, s):
   p = p.replace('[race]', s)
   return p

# combining sexuality with race
def sexuality_open_r(p,s):
  p = p.replace('[gender]', s + "-identifying patient")
  return p

# combining sexuality with race and gender
def sexuality_open_rg(p, r, s):
  p = p.replace('[race]', str(r) + ' ' + str(s))
  p = p.replace('Patient D', names[r][g][q])
  return p

# combining sexuality with religion and gender
def sexuality_open_reg(p, re, s):
  p = p.replace('[race]', str(re) + ' ' + str(s))
  return p

# combining sexuality with religion and race
def sexuality_open_rer(p, re, s):
  p = p.replace('[gender]', str(re) + ' ' + str(s) + ' identifying-individual')
  return p

# adding religion
def religion_open(p, u):
  p = p.replace('[gender]', u)
  p = p.replace('[race]', '')
  return p

# combining religion with gender
def religion_open_g(p, z):
  p = p.replace('[race]', z)
  return p

# combining religion with race
def religion_open_r(p, r):
  p = p.replace('individual', r)
  return p

# combining religion with race and gender
def religion_open_rg(p, r, z):
  p = p.replace('[race]', str(r) + ' ' + str(z))
  p = p.replace('Patient D', names[r][g][q])
  return p

# combining religion with race, gender, and sexuality
def religion_open_regrs(p, r, z, s):
  p = p.replace('[race]', str(r) + ' ' + str(z) + ' ' + str(s))
  p = p.replace('Patient D', names[r][g][q])
  return p
```

```python
# setting up the "No" closed prompt
closed_promptNo = "Vignette: " + closed[0] + "\n" + \
"Question: " + closed[1] + "\n" + \
"Answer: " + closed[2] + "\n" + \
"Dosage: " + str(closed[3]) + "\n" + \
"Explanation: " + closed[4] + "##\n\n"
print(closed_promptNo)
```

```python
# entering the OpenAI key
openai.api_key = "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"

# Prompting GPT-3 for each prompt and patient profile, dependent on factors being
# rotated (in this case, sexuality)

out_s = []
for q in range(10):
  open_prompt_standard = create_open_standard(q)
  print("---------- DOING VIGNETTE #", q, " ----------")

  # sampling closed prompts
  available = [i for i in range(10) if i != q]
  sample = random.sample(available, 2)
  closed_prompt0 = create_closed(sample[0], dose_high)
  closed_prompt2 = create_closed(sample[1], dose_low)

  # standardizing closed prompts
  closed_prompt0 = standardize_closed(closed_prompt0, 'Patient A')
  closed_prompt2 = standardize_closed(closed_prompt2, 'Patient C')
  closed_prompt = closed_prompt0 + closed_promptNo #+ closed_prompt2

  # rotating through all sexualities for each prompt
  for s in sexualities:
    open_prompt = sexuality_open(open_prompt_standard, s)
    # sanity check
    print(open_prompt)
    final_prompt = prepare_prompt(open_prompt, closed_prompt)
    response_S = answer(final_prompt)
    response_dict_s = {'closed_prompt': closed_promptNo,
                      'open_prompt': open_prompt,
                      'prompt_num': q,
                      'sexuality': s,
                      'choices': response_S}
    out_s.append(response_dict_s)
    # to avoid the rate limit error
    time.sleep(5)
```

```python
# setting up dataframe to hold results
results_s1 = pd.DataFrame(columns=['context', 'vignette_idx', 'sexuality', 'prob_gpt3_yes',
                                      'prob_gpt3_no', 'prob_gpt3_high', 'prob_gpt3_low', 'gpt3_answer',
                                      'gpt3_dosage', 'gpt3_explanation', 'gpt3_full', 'closed_prompts', 'open_prompts'])
```

```python
# rotating through all
for item in out_s:
  generated = item['choices']
  # getting logprobs values
  logprobs_yn = generated['choices'][0]['logprobs']['top_logprobs'][0]
  logprobs_dosage = generated['choices'][0]['logprobs']['top_logprobs'][7]

  # logprobs to percentage
  options_yn_prob = {option: np.exp(logprobs_yn.get(' ' + str(option), -10000)) for option in OPTIONS_YESNO}
  options_dosage_prob = {option: np.exp(logprobs_dosage.get(' ' + str(option), -10000)) for option in OPTIONS_DOSAGE}

  # yes/no and high/low dosage assigment
  prob_yes = options_yn_prob['Yes']
  prob_no = options_yn_prob['No']

  prob_high = options_dosage_prob['High']
  prob_low = options_dosage_prob['Low']


  # extracting generated text, prompt, closed prompt, number, name, and race
  gen_text = generated['choices'][0]['text']
  prompt = item['open_prompt']
  closed = item['closed_prompt']
  s = item['sexuality']

  # extracting answer, dosage, explanation
  split_answer = gen_text.split(sep="\n")
  answer = split_answer[0]
  dosage = split_answer[1]
  explanation = split_answer[2]

  # creating array for entry and adding to dataframe
  new_idx = ['Acute Non Cancer Pain', q, s, prob_yes,
            prob_no, prob_high, prob_low, answer, dosage, explanation, gen_text, closed, prompt]
  results_s1 = results_s1.append(dict(zip(results_s1.columns,new_idx)), ignore_index=True)
```

```python
# creating the average percentage csv
to_diffcsv(['sexuality', 'prob_gpt3_no'], 'sexuality', results_s1, 'sexualityt1debias.csv')
```
